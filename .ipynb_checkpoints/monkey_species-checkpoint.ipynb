# 10 Monkey Species

**Image dataset for fine-grain classification**

```python
%reload_ext autoreload
%autoreload
%matplotlib inline
```

```python
from fastai import *
from fastai.vision import *
```

## Looking at the data

```python
path = './data/monkeys'
```

```python
data = ImageDataBunch.from_folder(path, ds_tfms=get_transforms(), size=224)
```

**normalize the data**

```python
data.normalize(imagenet_stats)
```

**replace labels with actual class names**

```python
data.classes
```

```python
common_names = []

with open('data/monkey_labels.txt') as file:
    for i, line in enumerate(file):
        if i > 0 and i < 11:
            name = line.split(', ')[2].split()[0]
            common_names.append(name)
```

```python
common_names
```

```python
data.classes = common_names
```

```python
data.classes
```

**visualize the data**

```python
data.show_batch(rows=3, figsize=(7, 6))
```

# Training the model

```python
learn = ConvLearner(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(5)
```

## unfreeze the learner

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(3, max_lr=slice(1e-6, 1e-4))
```

## results

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9)
```

### confusion matrix

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

### most confused

```python
interp.most_confused(min_val=0)
```
